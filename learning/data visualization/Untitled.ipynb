```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# BASICS
x= np.linspace(0,5,11)
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title("cubic relation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# SUBPLOTS
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,y**2)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x**2,y)
```

```python
# OBJECT OREINTATED WAY
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.4,0.3,0.3])
axis2.plot(x,y)
```
